```python
from joblib import load

model = load('new_workflow.joblib')
```

```python
import pandas as pd

df = pd.read_csv("Datos_regresion.csv").drop("Diag.Egr1", axis=1)
X = df.drop(["Fallece"], axis=1)
y = df["Fallece"]
```

```python
preds = model.predict(
    df
)
```

```python
preds
```

```python
preds_prob = model.predict_proba(
    df
)
```

```python
preds_prob
```

```python
X["Fallece"] = y
X["Prediccion"] = preds
X["Prob Fallece"] = preds_prob[:,1]
```

```python
X[X['Fallece']!=X["Prediccion"]]
```

```python
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

print(confusion_matrix(y, preds))
print(classification_report(y, preds))
print(f"AUC-ROC: {roc_auc_score(y, preds):.2f}")
```

```python
X['Prob Fallece'] = round(X['Prob Fallece'],2) 
X.to_csv('data_with_pred_and_prob.csv')
```
